## Ensemble Metrics and Results

The following metrics and results are for the ensemble model:

- **Accuracy**: 0.947
- **Precision**: 0.943
- **Recall**: 0.951
- **F1 Score**: 0.947

The results dataframe contains the predictions, logits, and losses for each image in the test set.

```python
#Inporting the necessary libraries
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score 
import matplotlib.pyplot as plt
import cv2
```

```python
results = pd.read_csv('results/inference_ensemble.csv')
results['Prediction'] = results['Prediction'].apply(lambda x: int(x.strip('[]'))) # remove brackets
results['Logits'] = results['Logits'].apply(lambda x: float(x.strip('[]'))) # remove brackets

#Calculate the metrics

accuracy = accuracy_score(results['Target'], results['Prediction'])
precision = precision_score(results['Target'], results['Prediction'])
recall = recall_score(results['Target'], results['Prediction'])
f1 = f1_score(results['Target'], results['Prediction'])

print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1: {f1}')


#Have the images with the highest losses
results = results.sort_values(by='Loss', ascending=False)


#Display the top 10 images with the highest losses
print('Top 10 images with the highest losses')
for i in range(10):
    row = results.iloc[i]
    plt.figure(figsize=(2,2))
    plt.imshow(cv2.imread(row['File Path']))
    plt.title(f'Predicted: {row["Prediction"]}, Actual: {row["Target"]}')
    plt.show()
```
